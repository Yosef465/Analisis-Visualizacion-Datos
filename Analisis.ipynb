#### Valores atipicos

```python
valores = data.drop('ID', axis=1)

valores.head()
```

```python
columnas = valores.select_dtypes(include=['float64', 'int64']).columns

valores_atipicos = pd.DataFrame()
for column in columnas:
    Q1 = valores[column].quantile(0.25)
    Q3 = valores[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    atipicos = valores[(valores[column] < lower_bound) | (valores[column] > upper_bound)]
    valores_atipicos = pd.concat([valores_atipicos, atipicos])
    
print(valores_atipicos)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_columns = valores.select_dtypes(include=['float64', 'int64']).columns

sns.set(style="darkgrid")
num_plots = len(numerical_columns)
fig, axes = plt.subplots(nrows=num_plots, ncols=1, figsize=(8, 3 * num_plots))

if num_plots == 1:
    axes = [axes]

for i, column in enumerate(numerical_columns):
    sns.boxplot(data=data, x=column, ax=axes[i], color="red")
    axes[i].set_xlabel(column)

plt.tight_layout(pad=2.0)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# Definir las variables dependiente e independiente
X = data[['Ingreso anual_USD']]
Y = data[['Compra_USD']]

# Crear y entrenar el modelo de regresión lineal
model = LinearRegression()
model.fit(X, Y)
```

```python
coeficiente = model.coef_[0]
intercepto = model.intercept_

print(f'Coef: {coeficiente}')
print(f'Intercept: {intercepto}')
```

```python
plt.scatter(
    X, 
    Y,
    color='blue',
    label='Datos'
)

plt.plot(X, model.predict(X), color='red', linewidth=2, label='Regresion')
plt.title('Relacion entre ingreso anual y monto de compra')
plt.xlabel('Igreso anual')
plt.ylabel('Monto de Compra')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Agrupar los datos por 'Región' y calcular los promedios
agrupar_categoria = data.groupby('Región').agg({
    'Compra_USD': 'mean',
    'Frecuencia de compra': 'mean'
}).reset_index()

# Configurar el DataFrame para el mapa de calor
categorias = agrupar_categoria.set_index('Región')

# Crear la visualización del mapa de calor
plt.figure(figsize=(8, 6))
sns.heatmap(
    categorias,
    annot=True,
    cmap='YlGnBu',
    linewidths=0.5,
    fmt='.2f'
)

plt.title('Mapa de calor del Comportamiento por región')
plt.show()
```

# Modelo K-means
### identificar grupos de clientes

```python
print(data.columns)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans


# Selección de variables numéricas para el clustering
df_segmentacion = data[['Edad', 'Ingreso anual_USD', 'Compra_USD', 'Puntos de lealtad', 'Frecuencia de compra']]

# Escalado de las variables
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_segmentacion)

# Determinar el número óptimo de clústeres usando el Método del Codo
inercia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_scaled)
    inercia.append(kmeans.inertia_)

# Graficar el Método del Codo
plt.plot(range(1, 11), inercia, marker='o')
plt.xlabel('Número de Clústeres')
plt.ylabel('Inercia')
plt.title('Método del Codo')
plt.show()


```

```python
km = KMeans(n_clusters = 4)
```

```python
y_predict = km.fit_predict(df_segmentacion[['Edad', 'Ingreso anual_USD', 'Compra_USD', 'Puntos de lealtad', 'Frecuencia de compra']])
```

```python
y_predict
```

```python
df_segmentacion['cluster'] = (y_predict+1)
df_segmentacion.head()
```

```python
df_segmentacion.cluster.unique()
```

```python
print("Las coordenadas de los centroides de los cluster son: ")
km.cluster_centers_
```

```python
# Obtener las coordenadas de los centroides del modelo
centroides = km.cluster_centers_

# Gráfico de dispersión de los datos
plt.scatter(df_segmentacion['Ingreso anual_USD'], df_segmentacion['Compra_USD'], c=y_predict, cmap='rainbow')

# Agregar los centroides al gráfico
plt.scatter(centroides[:, 1], centroides[:, 2], marker='x', s=200, c='black', label='Centroides')

# Personalizar el gráfico
plt.title('Clusters y Centroides')
plt.xlabel('Ingreso anual (USD)')
plt.ylabel('Compra (USD)')
plt.legend()

plt.show()
```
